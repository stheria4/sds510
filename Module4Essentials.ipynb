This script imports the necessary libraries.

```python
import requests
from bs4 import BeautifulSoup
import csv
import time
```

This script loads a Mastodon hashtag page, looks for a keyword in each post, and saves any matches to a CSV file.

```python
# going to use a generic word to get some hits.
URL = "https://news.asu.edu/"
KEYWORD = "the"
CSV_FILE = "osint_hits.csv"

resp = requests.get(URL)
soup = BeautifulSoup(resp.text, "html.parser")

# find posts (Mastodon uses <article>)
posts = soup.find_all("article")

# Open CSV and write header
with open(CSV_FILE, "a", newline="", encoding="utf-8") as f:
    w = csv.writer(f)
    if f.tell() == 0:
        w.writerow(["keyword", "snippet", "url"])

    for post in posts:
        text = post.get_text(" ", strip=True)
        if KEYWORD.lower() in text.lower():
            snippet = text[:200]
            w.writerow([KEYWORD, snippet, URL])

print("Done.")
```
